```python
# Celda 1 — Importar librerías
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**3 - 3*x + 1

x  = np.linspace(-3, 3, 500)
y  = f(x)
dy = np.gradient(y, x)   # derivada numérica con numpy
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 7), sharex=True)

ax1.plot(x, y,  color='#5C6BC0', linewidth=2, label=r'$f(x) = x^3 - 3x + 1$')
ax1.axhline(0, color='gray', linewidth=0.8, linestyle='--')
ax1.set_ylabel('f(x)')
ax1.set_title('Función y su derivada')
ax1.legend()
ax1.grid(True, alpha=0.3)

ax2.plot(x, dy, color='#E53935', linewidth=2, label=r"$f'(x) = 3x^2 - 3$")
ax2.axhline(0, color='gray', linewidth=0.8, linestyle='--')
ax2.set_xlabel('x')
ax2.set_ylabel("f'(x)")
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
