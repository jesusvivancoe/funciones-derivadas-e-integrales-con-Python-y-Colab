**# Funciones en Python y Colab**

En esta sección vamos a estudiar funciones matemáticas y cómo representarlas gráficamente usando Python.

Consideremos la función cuadrática:

`f(x) = x^2 - 4x + 3`

Vamos a:
1. Evaluar la función en varios valores de \(x\).
2. Graficar su comportamiento.
---

```python
# Definimos la función en Python
def f(x):
    return x**2 - 4*x + 3

# Evaluamos la función en algunos valores
valores_x = [0, 1, 2, 3, 4, 5]
valores_y = [f(x) for x in valores_x]

print("x:", valores_x)
print("f(x):", valores_y)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generamos un rango continuo de valores de x
x = np.linspace(-1, 6, 100)
y = f(x)

# Graficamos la función
plt.plot(x, y, label=r'$f(x) = x^2 - 4x + 3$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfica de la función cuadrática')
plt.legend()
plt.grid(True)
plt.show()
```
